PSO paling baik digunakan untuk menemukan maksimum atau minimum dari suatu fungsi yang didefinisikan pada ruang vektor multidimensi. Asumsikan kita memiliki fungsi f (x) yang menghasilkan nilai nyata dari parameter X vektor (seperti koordinat (x,y) di pesawat) dan X dapat mengambil hampir semua nilai dalam ruang (misalnya, f(X) adalah ketinggian dan kita dapat menemukan satu untuk setiap titik di pesawat ), maka kita dapat menerapkan PSO. Algoritma PSO akan mengembalikan parameter X yang ditemukan yang menghasilkan minimum f(x).

```python
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
# Pengaturan hyperparameters
options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}
# memanggil contoh Particle Swarm Optimization (PSO)
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options)
# lakukan optimasi
best_cost, best_pos = optimizer.optimize(fx.sphere, iters=100)
```

```python
# riwayat biaya
optimizer.cost_history
# riwayat posisi
optimizer.pos_history
# riwayat kecepatan
optimizer.velocity_history
```

Modul pyswarms.utils.search berfungsi untuk mengimplementasikan berbagai teknik dalam pengoptimalan nilai hyperparameter dan modul pyswarms.utils.function mengimplementasikan berbagai fungsi pengujian untuk pengoptimalan

single_obj: kumpulan fungsi tujuan tunggal setiap baris mewakili partikel, dan setiap kolom mewakili posisinya pada dimensi tertentu dari ruang pencarian.

Di Sini, c1 dan c2adalah parameter kognitif dan sosial masing-masing. Mereka mengontrol perilaku partikel dengan dua pilihan: (1) mengikuti yang terbaik pribadinya atau (2) mengikuti posisi terbaik global kawanan . Secara keseluruhan, ini menentukan apakah gerombolan itu bersifat eksploratif atau eksploitatif. Selain itu, parameterw mengontrol inersia gerakan kawanan itu.

Untuk memplot riwayat biaya, kita cukup mendapatkan cost_historydari optimizerkelas dan meneruskannya ke cost_historyfungsi.

```python
import numpy as np
import pyswarms as ps
from pyswarms.utils.search import RandomSearch
from pyswarms.utils.functions import single_obj as fx

# pengaturan parameter
options = {
    'c1': (1,5),
    'c2': (6,10),
    'w': (2,5),
    'k': (11, 15),
    'p': 1
}

# buat random pencarian acak
# n_selection adalah jumlah iterasi untuk menjalankan pencarian
# iters adalah jumah iterasi untuk menjalankan pengoptimalan
g = RandomSearch(ps.single.LocalBestPSO, n_particles=40,
            dimensions=20, options=options, objective_func=fx.sphere,
            iters=10, n_selection_iters=100)

best_score, best_options = g.search()
```

Visualisasi Swarm
Dimungkinkan juga untuk memplot kinerja pengoptimal demi pemformatan. Modul plotters dibangun di atas matplotlib, membuatnya sangat dapat disesuaikan.
Dari code diatas dapat kita lihat jika iterasi dilakukan sebanyak 100 kali
Kemudian mengembalikan skor terbaik yang ditemukan selama pengoptimalan, dan opsi hyperparameter yang mengaktifkannya.

Langkah pertama adalah membuat pengoptimal. Di sini, kita akan menggunakan PSO Global-best untuk menemukan minimal fungsi sphere. Seperti biasa, kita cukup membuat instance kelasnya pyswarms.single.GlobalBestPSOdengan melewatkan parameter yang diperlukan yang akan kita gunakan. Kemudian, kita akan memanggil optimize()metode untuk 100 iterasi.

```python
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
from pyswarms.utils.plotters import plot_cost_history, plot_contour, plot_surface
import matplotlib.pyplot as plt
#pengaturan pengoptimalan
options = {'c1':0.5, 'c2':0.3, 'w':0.9}
optimizer = ps.single.GlobalBestPSO(n_particles=50, dimensions=2, options=options)
optimizer.optimize(fx.sphere, iters=100)
# plot
plot_cost_history(optimizer.cost_history)
plt.show()
```

```python
from pyswarms.utils.plotters.formatters import Mesher, Designer
# Plot the sphere untuk plot yang lebih baik
m = Mesher(func=fx.sphere,
           limits=[(-1,1), (-1,1)])
# sesuaikan batas gambar
d = Designer(limits=[(-1,1), (-1,1), (-0.1,1)],
             label=['x-axis', 'y-axis', 'z-axis'])
```

Modul plotters memiliki dua metode animasi: plot contour()dan plot surface(). Pendekatan ini, seperti namanya, memplot partikel dalam ruang 2-D atau 3-D.

Untuk memplot fungsi sphere, kita harus menambahkan mesh ke swarm kita. Hal ini memungkinkan kita untuk melihat di mana partikel berada dalam kaitannya dengan fungsi tujuan kita secara grafis.

Selain Mesher, ada kelas Designer untuk memodifikasi ukuran font, ukuran gambar, dan sebagainya, serta kelas Animator untuk mengatur penundaan dan pengulangan animasi.

plot contour() digunakan untuk memplot partikel dalam ruang 2D

```python
plot_contour(pos_history=optimizer.pos_history, mesher=m, designer=d, mark=(0,0))
```

plot surface() digunakan untuk memplot partikel dalam ruang 3D

```python
pos_history_3d = m.compute_history_3d(optimizer.pos_history) # preprocessing
animation3d = plot_surface(pos_history=pos_history_3d,
                           mesher=m, designer=d,
                           mark=(0,0,0))  
```
